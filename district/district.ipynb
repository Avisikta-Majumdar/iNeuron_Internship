# Now will build ML model

```python
from sklearn.cluster import KMeans
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler() 
```

```python
x= scaler.fit_transform(data[["A4","A5","A6","A7","A8","A9","A10","A11","A12","A13","A14","A15","A16"]])
```

```python
Data = pd.DataFrame(x , columns = ["A4","A5","A6","A7","A8","A9","A10","A11","A12","A13","A14","A15","A16"])
```

```python
Data
```

```python
DATA = Data.merge(data[["A3_Prague","A3_central Bohemia","A3_east Bohemia","A3_north Bohemia",
                 "A3_north Moravia","A3_south Bohemia","A3_south Moravia","A3_west Bohemia"]],left_index=True,
    right_index=True)
```

```python
DATA.head()
```

```python
#x_train, x_+test , y_traion , y_test = train_test_split()
```

```python
WCSS=[]
for i in range(1,11):
    Kmean = KMeans(n_clusters=i,)
    Kmean.fit(DATA)
    WCSS.append(Kmean.inertia_)
```

```python
WCSS
```

```python
plt.figure(figsize=(8,8))
plt.plot(np.arange(1,11),WCSS)
plt.xlabel("No Of Clusters",fontsize=15)
plt.ylabel("Wcss Value",fontsize=15)
plt.title("Wcss Vs. No. Of Clusters",fontsize=20)
```

```python
KMeans= KMeans(n_clusters=8)
KMeans.fit(DATA)
```

```python
y_pred = KMeans.fit_predict(DATA)
y_pred
```

```python
DATA["Cluster"] = y_pred
```
